```python
import matplotlib.pyplot as plot
import numpy as np
import numpy.fft as fft
%matplotlib inline 

a = 2.0
N = 10000
x = np.linspace(0, a, N)
fx = np.sin(3.75*2*np.pi*x / a  ) + np.sin(12.5*2*np.pi*x / a  )
plot.plot(x,fx)
plot.show()
```

```python
def fxfourier(œâ):
    N = len(fx)
    n = np.arange(N)
    a =(2/N)*1/(2*np.pi)**(1/2)
    M=a*np.exp(-2j * np.pi * œâ * n / N)
    return np.dot(M, fx)


frange=range(0,200)
freqf=[fxfourier(0.1*x) for x in frange]
```

```python
plot.plot(0.1*np.asarray(frange),np.abs(freqf))
```
